```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis toa list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pyber Fare by Month")
# Add the legend.
plt.legend()
plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="*", linewidth=2)
ax.legend("Boston")
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 50)
ax.grid()
plt.show()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create the labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Create a title.
plt.title("Pyber Fare by month")
# Add a legend.
plt.legend()
plt.show()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("Pyber Fare by Month")
ax.invert_yaxis()
plt.show()
```

```python
# Create the plot - using MATLAB method.
plt.plot(x_axis, y_axis, "o")
plt.show()
```

```python
# Alternate way of creating the scatter plot with MATLAB.
plt.scatter(x_axis, y_axis)
plt.show()
```

```python
# Annottating scatter plot.
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Creating Bubble chart - set size of markers equal to a feature of our data.
plt.scatter(x_axis, y_axis, s=y_axis)
plt.show()
```

```python
# To highlight differences in points of our data, we can adjust size
# by multiplying each marker by 3.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data * 3)
plt.scatter(x_axis, y_axis, s = y_axis_larger)
plt.show()
```

```python
# Adjusting the size of markers on Bubble chart using list comprehensions.
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
plt.show()
```

```python
# Create the scatter plot using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
plt.show()
```

```python
# Create a Bubble Chart using the Object-Oriented approach.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

```python
# Create a Bubble Chart using the Object-Oriented approach.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha=.8,
          edgecolors="k", linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
plt.show()
```

```python
# Create Pie Chart - using MATLAB method.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Exploding a slice of the pie and adding percentages to the wedges.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create Pie Chart using object oriented method.
explode2 = [0, 0, 0.2, 0, 0, 0, .2, 0, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
       labels=x_axis, 
      explode=explode2, 
      shadow=True,
      startangle=95,
      counterclock=False,
      colors=colors,
      autopct="%.1f%%")
plt.show()
```
